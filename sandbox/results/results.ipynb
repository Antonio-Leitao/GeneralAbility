```python
import numpy as np
```

```python
seeds = range(30)
```

```python
print('\t|', 'Seed\t','|', ' Train Mae\t', '|', ' Test Mae\t', '|', ' Gen Score\t', '|', '  P Score\n')

epoch = -1
custom = []
control = []

for seed in seeds:

    temp = np.load(f'{seed}_custom.npy', allow_pickle=True)
    
    custom.append([np.round(temp.item()['gen_score'][epoch],4), np.round(temp.item()['test_mae'][epoch],4)])
    
    print('Custom',
          '\t    ', 
          seed,  
          '\t    ', 
          np.round(temp.item()['mae'][epoch],4),
          '\t    ', 
          np.round(temp.item()['test_mae'][epoch],4), 
          '\t    ',
          np.round(temp.item()['gen_score'][epoch],4),  
          '\t    ', 
          np.round(temp.item()['p_score'][epoch],4))

    
    temp = np.load(f'{seed}_control.npy', allow_pickle=True)
    
    control.append([np.round(temp.item()['gen_score'][epoch],4), np.round(temp.item()['test_mae'][epoch],4)])
    
    print('Control     ', 
          seed,  
          '\t    ', 
          np.round(temp.item()['mae'][epoch],4), 
          '\t    ', 
          np.round(temp.item()['test_mae'][epoch],4), 
          '\t    ', 
          np.round(temp.item()['gen_score'][epoch],4), 
          '\t    ', 
          np.round(temp.item()['p_score'][epoch],4))
    
    print('\n')
```

```python
print('\t Generalization Score:\n')

print('Custom \t\t', np.median(np.array(custom)[:,0]).round(4))

print('Control \t', np.median(np.array(control)[:,0]).round(4))

print('\n', '\t\t Test Mae:\n')

print('Custom \t\t', np.median(np.array(custom)[:,1]).round(4))

print('Control \t', np.median(np.array(control)[:,1]).round(4))
```
